# Stocks Analysis

This is an analysis report which depicts the trends of stocks associated with major companies like Amazon, Apple, Neflix and Tesla. I have used datasets for all these companies which contains their overall stocks effective of 2016-2021(present). In this analysis, I will be demonstorating the changes which occured due to COVID-19 and the factors responsible for those changes. 

We will first be looking into a basic line graph which gives us a good visual of how stock trends look like. As we can clearly see, Amazon with the blue line has highest stock values as compared to other companies. COVID-19's first case was reported on December 19th 2019 in the United States. From the graph, we can clearly visualize that Amazon stocks immediately went really high soon after the first reported case in December 2019. Tesla on the other hand had a very minor increase in stock. Netflix and Apple are reported to have no drastic changes after December 2019.

The second graph attempts to show the daily volume changes around the announcement of COVID-19.

Let's look into how the exact numbers look like. The stocks before covid contains the open stock price effective of January 2016 - December 2019. Stocks during covid contains the dataset from January 2020- October 2021 (present). 

As we can clearly see, Amazon stock price nearly doubled after the hit of pandemic. Tesla on the other hand had a huge dranstic increase in stocks right after pandemic hit. Netflix's stock price also increased nearly two times. Apple also saw a big shift in their stock price. 

These numbers seem reasonable since 2020 was the peak of COVID-19 and with sudden lockdown, it was nearly impossible for people to shop in-person for all their daily needs. In today's time, we can almost buy anything online thorugh the amazon app in the comfort of our homes and with amazing delivery services, Amazon became more and more useful for all of us. Netflix on the other hand was very useful for entertainment purposes. People were unable to step out to refresh their minds and Netflix was one of the other most useful applications to survive through global pandemic. Talking about Apple, with almost everything switching virtually, everyone required a desktop or a laptop to continue working as they used to before the pandemic. Even small kids were attending their school online and with that, almost everyone required their personal laptop or a desktop device. Last but not the least, we will be looking into the factors associated with Tesla's big shift later in this document.

```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame

df=pd.DataFrame()

amazon_before=pd.read_csv('Amazon.csv')
tesla_before=pd.read_csv('TSLA.csv')
netflix_before=pd.read_csv('netflix.csv')
apple_before=pd.read_csv('Apple.csv')

df['Company']=['Amazon', 'Tesla', 'Netflix', 'Apple']
df['Stocks before covid']=[amazon_before['Open'].iloc[4692:5688].mean(),tesla_before['Open'].iloc[1:58].mean(),netflix_before['Open'].iloc[3429:4426].mean(),apple_before['Open'].iloc[8843:9840].mean()]
df['Stocks during covid']=[amazon_before['Open'].iloc[5688:].mean(),tesla_before['Open'].iloc[58:].mean(),netflix_before['Open'].iloc[4426:].mean(),apple_before['Open'].iloc[9840:].mean()]
df
```
